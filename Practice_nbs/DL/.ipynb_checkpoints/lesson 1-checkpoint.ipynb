# data and model initlisation 

```python
data=ImageDataBunch.from_name_re(path,
                                 valid_pct=.3,
                                 size=224,
                                 bs=16,
                                 fnames=fnames,
                                 pat=r'/([^/]+)_\d+.jpg',
                                ds_tfms=get_transforms()).normalize(imagenet_stats)
```

```python
len(data.classes),len(data.train_ds),len(data.valid_ds)
```

```python
data.classes[19]
```

```python
data.show_batch(rows=1,figsize=(6,5))
```

```python
learner=cnn_learner(data,models.resnet34,metrics=error_rate)
```

```python
learner.lr_find()
```

```python
learner.recorder.plot(suggestion=True)
```

```python
learner.fit_one_cycle(4,3e-03)
```

# res50 

```python
data=ImageDataBunch.from_name_re(path,
                                 valid_pct=.3,
                                 size=512,
                                 bs=16,
                                 fnames=fnames,
                                 pat=r'/([^/]+)_\d+.jpg',
                                ds_tfms=get_transforms()).normalize(imagenet_stats)
```
